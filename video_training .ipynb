```python
accelerator = pd.DataFrame(info.get('accelerometer'))
accelerator.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


for colum_data in accelerator.T.iteritems():
    ax.scatter(colum_data[1].get('x'), colum_data[1].get('y'), colum_data[1].get('z'))
    #print(colum_data[1])
    

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
